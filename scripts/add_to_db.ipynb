```python
import requests
import sqlite3
```

```python
def add_to_db(number, poem, interpretation):

    url = "https://hafezdivination.pythonanywhere.com/api/"
    
    body = {
        "number": number,
        "poem": poem,
        "interpretation": interpretation,
        "is_active": True
    }

    response = requests.post(url, json=body)

    print(response.text)
```

```python
database_path = '../db.sqlite3'
conn = sqlite3.connect(database_path)
cursor = conn.cursor()

cursor.execute("SELECT * FROM poems")
rows = cursor.fetchall()
```

```python
for row in rows:

    if row[6] <= 259:
        continue

    add_to_db(
        row[6],
        row[1],
        row[2]
    )
```
