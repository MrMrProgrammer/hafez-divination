```python
import requests
import sqlite3
```

```python
def add_to_db(number, poem, interpretation):

    url = "http://127.0.0.1:8000/poem/"
    
    body = {
        "number": number,
        "poem": poem,
        "interpretation": interpretation,
        "is_active": True
    }

    response = requests.post(url, json=body)

    print(response.text)
```

```python
database_path = 'db.sqlite3'
conn = sqlite3.connect(database_path)
cursor = conn.cursor()

cursor.execute("SELECT * FROM mainApp_poems")
rows = cursor.fetchall()
```

```python
for row in rows:
    add_to_db(
        row[0],
        row[1],
        row[2]
    )
```
